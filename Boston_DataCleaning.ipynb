```python
#DATA CLEANING (Google-Style, Industry Standard)
import pandas as pd
import numpy as np

df = pd.read_csv("/content/Somerville Crashes data.csv")
df_clean = df.copy()
```

```python
df_clean.columns =(
    df_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(" ","_")
)
df_clean.columns
```
