```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/content/startup_data.csv")
df
```

```python
df.columns = (
    df.columns
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False)
    .str.strip()
    .str.lower()                         # remove extra spaces
    .str.replace(" ", "_")
)

df.columns
```

```python
missing_values = df.isnull().sum()
print("Missing Values Per Columns")
print(missing_values)
```

```python
print("Duplicated Row Count :", df.duplicated().sum())
```

```python
cat_cols = df.select_dtypes(include = "object").columns.tolist()
num_cols = df.select_dtypes(include =["int64", "float64"]).columns.tolist()


print("Categorial Columns: ",cat_cols)
print("Numerical Columns :", num_cols)


print("\n Unique values and their Counts for categorial Columns: ")
for col in cat_cols:
  print("Values Count for {col} :")
  print(df[col].value_counts())
```

```python
print("Datatypes in each columns:")
print(df.dtypes)
```

```python
industry_distribution = df['industry'].value_counts()
print("Distribution of Startups across Industries")
print(industry_distribution)
```

```python
plt.figure()
df['industry'].value_counts().plot(kind = "bar")
plt.title("Distribution of Startups across Industries")
plt.xlabel("Industries")
plt.ylabel("Count of Startups")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
industry_distribution.plot(kind = 'bar')
plt.title('Distribution of Startups across Industries')
plt.xlabel('Industries')
plt.ylabel('Number of Startups')
plt.xticks(rotation=45 , ha ='right')
plt.tight_layout()
plt.show()
```

```python
df.groupby('industry')['valuation_m_usd'2].mean().sort_values(ascending = True)
```

```python
avg_valuation = df.groupby("industry")["valuation_m_usd"].mean()
avg_funding = df.groupby("industry")["funding_amount_m_usd"].mean()
print("Average Funding by Industry")
print(avg_funding)


print("Average Valuatuion by Industry")
print(avg_valuation)
```

```python
#Plot for Average Funding amount for Industries
plt.figure(figsize= (12,7))
avg_funding.plot(kind = "bar")
plt.title("Average Funding Amount Across Industries")
plt.xlabel("Industry")
plt.ylabel("Average Funding Amount")
plt.xticks(rotation= 45, ha = 'right')
plt.tight_layout()
plt.show()


#Plot for Average Valuation amount for Industries
plt.figure(figsize= (12,7))
avg_valuation.plot(kind = "bar")
plt.title("Average Valuation Amount across Industries")
plt.xlabel("Industry")
plt.ylabel("Average Funding Amount")
plt.xticks(rotation = 45, ha= 'right')
plt.tight_layout()
plt.show()
```

```python
avg_revenue = df.groupby("industry")["revenue_m_usd"].mean()
avg_marketshare = df.groupby("industry")["market_share_%"].mean()


print("Average Revenue by Industry")
print(avg_revenue)


print("Average Market Shares by Industry")
print(avg_marketshare)
```

```python
#Average Revenue Across Industry
plt.figure(figsize= (12,7))
avg_revenue.plot(kind= "bar", color = "darkgreen")
plt.title("Average Revenue Across Industry")
plt.xlabel("Industry")
plt.ylabel("Average Revenue")
plt.xticks(rotation = 45 , ha = 'right')
plt.tight_layout()
plt.show()


#Average Market Share of Industries
plt.figure(figsize = (12,7))
avg_marketshare.plot(kind= "bar", color ="pink")
plt.title("Average MArket Share Across Industries")
plt.xlabel("Industry")
plt.ylabel("Average Market Share")
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

```python
profitability = df.groupby("industry")["profitable"].mean()
print("Profitability by Industry")
print(profitability)
```

```python
exitstatus = df.groupby('industry')['exit_status'].value_counts().unstack(fill_value =0)
print("Exit Status across Industries")
print(exitstatus)
```

```python
#Profitability across Industries
plt.figure(figsize= (12,7))
profitability.plot(kind = "bar", color = "teal" )
plt.title("Profitability Across Industries")
plt.xlabel("Industry")
plt.ylabel("Avergae Profitablity Count")
plt.xticks(rotation = 45, ha= 'right')
plt.tight_layout()
plt.show()


#Exits status of Industries
plt.figure(figsize= (12,7))
exitstatus.plot(kind= "bar", stacked = True , cmap = "viridis")
plt.title("Exit Status Across Industries")
plt.xlabel("Industry")
plt.ylabel("Exit Status Count")
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

```python
emp_count = df.groupby('industry')['employees'].mean()
print("Employee Count by Industry")
print(emp_count)
```

```python
#Industries employee count
plt.figure(figsize = (12,7))
emp_count.plot(kind='bar', color = 'skyblue')
plt.title("Industries Employee Count")
plt.xlabel("Industry")
plt.ylabel("Employee Number")
plt.xticks(rotation = 45 , ha='right')
plt.tight_layout()
plt.show()
```

```python
avg_year_founded = df.groupby('industry')['year_founded'].mean()
print("Average Year Founded by Industry")
print(avg_year_founded)
```

```python
#verage Year Founded by Industry
plt.figure(figsize = (12,7))
avg_year_founded.plot(kind='bar', color = 'skyblue')
plt.title("Average Year Founded by Industry")
plt.xlabel("Industry")
plt.ylabel("Average Year")
plt.xticks(rotation = 45 , ha='right')
plt.tight_layout()
plt.show()
```
