```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Customer_Sentiment.csv")
df
```

```python
missing_values = df.isnull().sum()
print("\n Missing Values Per Column :")
print(missing_values)
```

```python
duplicates = df.duplicated().sum()
print("Duplicated Row Count: ")
print(duplicates)
```

```python
age_distribution = df['age_group'].value_counts()
print(age_distribution)
```

```python
#age-WISE SHOPPING DISTRIBUTION
plt.figure(figsize = (12, 7))
age_distribution.plot(kind = 'bar', color ='lightgreen')
plt.title('Age-wise Shopping Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

```python
gender_age_distribution = df.groupby('gender')['age_group'].value_counts()
print(gender_age_distribution)
```

```python
#gender_age_distribution
plt.figure(figsize = ( 12 ,7 ))
gender_age_distribution.plot(kind = 'bar', color= 'teal')
plt.title('Gender & Age-wise Shopping Distribution')
plt.xlabel('Gender-age Grouping')
plt.ylabel('Count')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
sentiment_analysis= df['sentiment'].value_counts()
print(sentiment_analysis)
```

```python
#SENTIMENT ANALYSIS
plt.figure(figsize = (12, 7))
sentiment_analysis.plot(kind = 'bar')
plt.title("Overall Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Number of Customers")
plt.xticks(rotation = 45 , ha = 'right')
plt.tight_layout()
plt.show()
```

```python
sentiments= df.groupby('product_category')['sentiment'].value_counts().unstack(fill_value = 0)
print(sentiments)
```

```python
#Sentiments by Product Category
plt.figure(figsize = (12, 7))
sentiments.plot(kind = 'bar',stacked = True, cmap = 'viridis')
plt.title('Sentiments by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Setiments Count')
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

```python
response_sentiments = df.groupby('response_time_hours')['sentiment'].value_counts()
print(response_sentiments)
```

```python
issues = df.groupby('complaint_registered')['issue_resolved'].value_counts().unstack(fill_value = 0)
print(issues)
```

```python
#compaint registered vs issues resolved
plt.figure(figsize = (12, 7))
issues.plot(kind = 'bar', stacked = True, cmap = 'viridis')
plt.title('compaint registered vs issues resolved')
plt.xlabel('Complaint Registered')
plt.ylabel('count of Issues Resolved')
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

```python
platform_sentiments = df.groupby('platform')['sentiment'].value_counts().unstack(fill_value = 0)
print(platform_sentiments)
```

```python
#PLATFORM VS SENTIMENTS
plt.figure(figsize = (12, 7))
platform_sentiments.plot(kind = 'bar', stacked = True, cmap = 'coolwarm')
plt.title('Platform vs Setiments')
plt.xlabel('Platforms')
plt.ylabel('Sentiments')
plt.xticks(rotation = 45, ha ='right')
plt.tight_layout()
plt.show()
```

```python
#PLATFORM VS SENTIMENTS
plt.figure(figsize = (12, 7))
platform_sentiments.plot(kind = 'bar')
plt.title('Platform vs Setiments')
plt.xlabel('Platforms')
plt.ylabel('Sentiments')
plt.xticks(rotation = 45, ha ='right')
plt.tight_layout()
plt.show()
```

```python
platform_issuesresolved = df.groupby('platform')['issue_resolved'].value_counts().unstack(fill_value = 0)
print(platform_issuesresolved)
```

```python
Platform_complaints = df.groupby('platform')['complaint_registered'].value_counts().unstack(fill_value = 0)
print(Platform_complaints)
```

```python
#platform vs complaints
plt.figure(figsize = ( 12, 7))
Platform_complaints.plot(kind = 'bar', color=['skyblue', 'red'])
plt.title('Platform vs complaints')
plt.xlabel('Platform')
plt.ylabel('Complaints')
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()


#platforms vs issues resolved
plt.figure(figsize = (12, 7))
platform_issuesresolved.plot(kind= 'bar')
plt.title('Platform vs Issuesresolved ')
plt.xlabel('Platforms')
plt.ylabel('Issues Resolved')
plt.xticks(rotation = 45 , ha = 'right')
plt.tight_layout()
plt.show()
```

```python
P_customer_ratings = df.groupby('platform')['customer_rating'].mean()
print(P_customer_ratings )
```

```python
#Platform vs Average Customer Ratings
plt.figure(figsize = (12,7))
P_customer_ratings.plot(kind ='bar', color = 'skyblue')
plt.title('Platform vs Average Customer Ratings')
plt.xlabel('Platforms')
plt.ylabel('Average Customer Ratings')
plt.xticks(rotation = 45 , ha = 'right')
plt.tight_layout()
plt.show()
```
