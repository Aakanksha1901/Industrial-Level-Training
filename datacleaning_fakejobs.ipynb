```python
isfake = df['is_fake'].value_counts(normalize = True) * 100
print(isfake)
```

```python
#Real vs Fake Distribution
plt.figure(figsize = (12,7))
isfake.rename({0: 'Real Job', 1:'Fake job'}).plot(kind = 'bar')
plt.title('Real vs Fake Job Distribution')
plt.xlabel('Real or Fake')
plt.ylabel('Percentile(%)')
plt.xticks(rotation = 45 , ha = 'right')
plt.tight_layout()
plt.show()
```

```python
havinglogo = df.groupby('has_logo')['is_fake'].value_counts().unstack(fill_value = 0)
print(havinglogo)
```

```python
plt.figure(figsize = (12,7))
havinglogo.rename({0: 'No' , 1:'Yes'}).plot(kind = 'bar', stacked = True)
plt.title ('Has Logo vs Fake Jobs')
plt.xlabel('Has Logo')
plt.ylabel('Count of Jobs')
plt.legend(title ='Is Fake (0 = No , 1 = Yes)')
plt.xticks(rotation = 45 , ha = 'right')
plt.show()
```

```python
iftelecommuting = df.groupby('telecommuting')['is_fake'].value_counts().unstack(fill_value = 0)
print(iftelecommuting)
```

```python
plt.figure(figsize = (12,7))
iftelecommuting.rename({0: 'Onsite', 1: 'Remote'}).plot(kind = 'bar' , color =['skyblue' , 'darkblue'])
plt.title('Telecommuting vs Fake Jobs')
plt.xlabel('Telecommuting')
plt.ylabel('Count of Jobs')
plt.xticks (rotation = 45 , ha = 'right')
plt.legend(title = 'is fake(0 = Real Job, 1 = Fake job)')
plt.tight_layout()
plt.show()
```

```python
employmenttype = df.groupby('employment_type')['is_fake'].value_counts().unstack(fill_value = 0)
print(employmenttype)
```

```python
industry_fake = df.groupby('industry')['is_fake'].value_counts().unstack(fill_value = 0)
print(industry_fake)
```

```python
fake_jobs = df[df['is_fake'] == 1]
fake_industry_count = fake_jobs['industry'].value_counts().sort_values(ascending = False)
print(fake_industry_count)
```

```python
plt.figure(figsize=(12,7))

real_exp = df[df['is_fake'] == 0]['required_experience_years'].dropna()
fake_exp = df[df['is_fake'] == 1]['required_experience_years'].dropna()

plt.boxplot(
    [real_exp, fake_exp],
    labels=['Real Jobs (0)', 'Fake Jobs (1)'],
    showmeans=True
)

plt.title('Fake Jobs by Required Experience Years')
plt.xlabel('Job Type')
plt.ylabel('Required Experience (Years)')
plt.show()
```

```python
df['Date']  = pd.to_datetime(df['posting_date'])
df['Year']  = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['YearMonth'] = df['Date'].dt.to_period('M')

monthly_counts = df.groupby('YearMonth').size()

print(monthly_counts)
```

```python
#Monthly Job Posting Trends
plt.figure(figsize = (12,7))
monthly_counts.plot(kind = 'line', color = 'red', marker = 'o')
plt.title('Monthly Job Posting Trends')
plt.xlabel('Year-Month')
plt.ylabel('Number of Job Postings')
plt.xticks(rotation = 45 , ha = 'right')
plt.tight_layout()
plt.show()
```

```python
trend_data = df.groupby(['YearMonth', 'is_fake']).size().unstack(fill_value=0)
print(trend_data)
```

```python
trend_data.index = trend_data.index.astype(str)
```

```python
#Real vs Fake Job Posting Trends
plt.figure(figsize = (12,7))
plt.plot(trend_data.index, trend_data[0], marker = 'o', label = 'Real Jobs', color = 'orange')
plt.title('Monthly Trend of Real Jobs')
plt.xlabel('Year - Month')
plt.ylabel('Number of Job Posting')
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()


#Real vs Fake Job Posting Trends
plt.figure(figsize = (12,7))
plt.plot(trend_data.index, trend_data[1], marker = 'o', label = 'Fake Jobs')
plt.title('Monthly Trend of Fake Jobs')
plt.xlabel('Year - Month')
plt.ylabel('Number of Job Posting')
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

```python
df['posting_date'] = pd.to_datetime(df['posting_date'])
df['application_deadline'] = pd.to_datetime(df['application_deadline'])
```

```python
df['deadline_days'] = (df['application_deadline'] - df['posting_date']).dt.days
```

```python
real_deadline = df[df['is_fake'] == 0]['deadline_days'].dropna()
fake_deadline = df[df['is_fake'] == 1]['deadline_days'].dropna()
```

```python
plt.figure(figsize=(8,6))

plt.boxplot(
    [real_deadline, fake_deadline],
    labels=['Real Jobs (0)', 'Fake Jobs (1)'],
    showmeans=True
)

plt.title('Application Deadline vs Fake Jobs')
plt.xlabel('Job Type')
plt.ylabel('Days Until Deadline')
plt.tight_layout()
plt.show()
```
