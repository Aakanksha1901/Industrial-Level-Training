```python
cat_cols = df.select_dtypes(include =['object']).columns.tolist()
num_cols = df.select_dtypes(include = ['float64']).columns.tolist()

print("Categorial Columns: ", cat_cols)
print("Numerical Columns: ",num_cols)

for col in cat_cols:
    print(f"\n Values count for '{col}': ")
    print(df[col].value_counts())
```

```python
print("Duplicated Row count:" ,df.duplicated().sum())
```

```python
df['product_id']   = df['product_id'].fillna('NA')
df['product_name'] = df['product_name'].fillna('NA')
df['category']     = df['category'].fillna('NA')
df['price'] = df['price'].fillna('NA')
df['reviews_count'] = df['reviews_count'].fillna('NA')
df['seller'] = df['seller'].fillna('NA')
df['stock'] = df['stock'].fillna(df['stock'].mean())
df['shipment_type'] = df['shipment_type'].fillna('NA')
df['order_date'] = df['order_date'].fillna('Invalid Date')
df['delivery_date'] = df['delivery_date'].fillna('Not a date')
df['discount_percent'] = df['discount_percent'].fillna('')
print(df.isnull().sum())
```
